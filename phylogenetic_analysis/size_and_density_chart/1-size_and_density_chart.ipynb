```python
# Coluna do tamanho dos genomas isolados em bp (pares de base)

## Manualmente
### Vamos utilizar este codigo para calcular o tamanho do genoma em bp de uma sequencia em fasta 
awk '/^>/ {next} {total += length($0)} END {print total}' SA10-UEL.fa

## Automatizado
### Criar o script contar_bases.sh
nano contar_bases.sh
#### Copiar e colar o código abaixo no arquivo contar_bases.sh ####

#####################################################################################################################################################
#!/bin/bash

# Diretórios a processar, sempre entre aspas e separados somente por espaço (sem vírgulas)
dirs=("genoma")

# Arrays para armazenar dados para o resumo
declare -A dir_counts
declare -A dir_totals

echo "Processando arquivos FASTA..."
echo "======================================"

# Loop pelos diretórios
for dir in "${dirs[@]}"; do
    echo ""
    echo "Processando diretório: $dir"
    echo "--------------------------------------"
    
    # Verifica se o diretório existe
    if [ ! -d "$dir" ]; then
        echo "AVISO: Diretório $dir não encontrado!"
        continue
    fi
    
    # Inicializa contadores
    dir_counts[$dir]=0
    dir_totals[$dir]=0
    
    # Loop pelos arquivos .fasta no diretório
    for file in "$dir"/*_genomic.fna; do
        # Verifica se existem arquivos
        if [ ! -f "$file" ]; then
            echo "Nenhum arquivo encontrado em $dir"
            break
        fi
        
        # Extrai apenas o nome do arquivo
        filename=$(basename "$file")
        
        # Executa o awk e captura o resultado
        total=$(awk '/^>/ {next} {total += length($0)} END {print total}' "$file")
        
        # Exibe o resultado
        echo "$dir/$filename: $total bases"
        
        # Atualiza contadores
        ((dir_counts[$dir]++))
        ((dir_totals[$dir]+=$total))
    done
done

echo ""
echo "======================================"
echo "Processamento concluído!"
echo ""
echo "Resumo por diretório:"
echo "--------------------------------------"

# Gera resumo por diretório
for dir in "${dirs[@]}"; do
    if [ -d "$dir" ]; then
        echo "$dir: ${dir_counts[$dir]} arquivos, ${dir_totals[$dir]} bases totais"
    fi
done
#####################################################################################################################################################
chmod +x contar_bases.sh

./contar_bases.sh
```

```python
# Ir para o diretório ISEScan
cd /mnt/dados/victor_baca/outputs/size_and_density_chart/ISEScan

# Criar o script para contar a quantidade e saber quais são as famílias de Insertion Sequences (IS) em arquivo FASTA

# Criar o script contar_familias.py
nano contar_familias.py
## Copiar e colar o código abaixo no arquivo contar_familias.py ##

#####################################################################################################################################################
#!/usr/bin/env python3

import re
from collections import Counter
import sys
import os

def parse_fasta_and_count_families(filename):
    """
    Lê arquivo FASTA e conta as famílias de IS encontradas
    
    Args:
        filename: caminho do arquivo FASTA
    
    Returns:
        Counter object com contagem das famílias
    """
    families = []
    
    try:
        with open(filename, 'r') as f:
            for line in f:
                # Processa apenas linhas de cabeçalho (começam com >)
                if line.startswith('>'):
                    # Procura pelo padrão family=NOME_FAMILIA
                    match = re.search(r'family=([^;]+)', line)
                    if match:
                        family_name = match.group(1)
                        families.append(family_name)
    
    except FileNotFoundError:
        print(f"ERRO: Arquivo '{filename}' não encontrado!")
        sys.exit(1)
    except Exception as e:
        print(f"ERRO ao ler arquivo: {e}")
        sys.exit(1)
    
    return Counter(families)

def main():
    # Verifica se o arquivo foi fornecido como argumento
    if len(sys.argv) != 2:
        print("USO: python3 contar_is.py <arquivo.fasta>")
        print("\nExemplo:")
        print("  python3 contar_is.py Bovine_cdhit.fasta")
        print("  python3 contar_is.py /caminho/completo/para/arquivo.fasta")
        sys.exit(1)
    
    filename = sys.argv[1]
    
    # Verifica se o arquivo existe
    if not os.path.isfile(filename):
        print(f"ERRO: Arquivo '{filename}' não encontrado!")
        print(f"Caminho verificado: {os.path.abspath(filename)}")
        sys.exit(1)
    
    print("="*60)
    print("ANÁLISE DE FAMÍLIAS DE INSERTION SEQUENCES (IS)")
    print("="*60)
    print(f"\nArquivo: {filename}")
    print(f"Caminho completo: {os.path.abspath(filename)}\n")
    
    # Faz a contagem
    family_counts = parse_fasta_and_count_families(filename)
    
    if not family_counts:
        print("Nenhuma família de IS encontrada no arquivo!")
        return
    
    # Mostra resultados
    print(f"Total de sequências IS encontradas: {sum(family_counts.values())}")
    print(f"Número de famílias diferentes: {len(family_counts)}\n")
    
    print("-"*60)
    print(f"{'FAMÍLIA':<30} {'QUANTIDADE':>15}")
    print("-"*60)
    
    # Ordena por quantidade (decrescente) e depois por nome
    for family, count in sorted(family_counts.items(), 
                                key=lambda x: (-x[1], x[0])):
        print(f"{family:<30} {count:>15}")
    
    print("-"*60)
    print(f"{'TOTAL':<30} {sum(family_counts.values()):>15}")
    print("="*60)

if __name__ == "__main__":
    main()
#####################################################################################################################################################

chmod +x contar_familias.py

python3 contar_familias.py Bovine/cd-hit-est/8B14M_isescan_cdhit.fasta
```

```python
cd /mnt/dados/victor_baca/outputs/size_and_density_chart/BLASTn

# Script  Para filtrar arquivos FASTA por TEs específicos, removendo os overlaps ja analisados
nano filter_script.py
# Copiar o código abaixo

#####################################################################################################################################################
#!/usr/bin/env python3

import os
import re
from pathlib import Path

# Lista de TEs para filtrar
TES = [
    'ISSag9','ISSag3','ISSag5','ISSag2','IS1381A','ISSag8','IS861','ISSag4','ISStin1','ISSag12','ISSpy1','ISCco2','IS1167','ISSag11','ISSmu1','IS1548','IS1216E','IS1167A','ISSg1','IS1550','ISSsu4','ISSth2','ISSa4','IS256','ISLgar1','IS1252','IS6770','IS1194','IS-LL6','IS981','ISLla2','ISSsu5','ISSdy1','ISSdy2','IS200S','ISPsy14','ISPst2','ISPpu17','ISShes10','ISSth1','ISScr1','ISStso1','ISSsu7','ISSsu6','IS1239','ISSpy2','ISSag10','ISLgar5','ISBun2','TnSsu5','Tn5531','Tn551','Tn1546','Tn6246','Tn7114','Tn2670','TnSs1','Tn6674','Tn4001','Tn5393','Tn501','Tn5041','Tn6214','Tn2'
]

# Subdiretórios a processar
SUBDIRS = ['Fish', 'Human', 'Bovine']

def header_contains_te(header, te_list):
    """
    Verifica se o cabeçalho contém algum dos TEs da lista.
    
    Args:
        header: String do cabeçalho FASTA
        te_list: Lista de nomes de TEs
        
    Returns:
        bool: True se contém algum TE, False caso contrário
    """
    header_upper = header.upper()
    for te in te_list:
        # Busca case-insensitive pelo nome do TE
        if te.upper() in header_upper:
            return True
    return False

def filter_fasta(input_file, output_file, te_list):
    """
    Filtra um arquivo FASTA mantendo apenas sequências com TEs específicos.
    
    Args:
        input_file: Caminho do arquivo FASTA de entrada
        output_file: Caminho do arquivo FASTA de saída
        te_list: Lista de nomes de TEs para filtrar
    """
    kept_sequences = 0
    total_sequences = 0
    
    with open(input_file, 'r') as fin, open(output_file, 'w') as fout:
        current_header = None
        current_sequence = []
        keep_sequence = False
        
        for line in fin:
            line = line.rstrip('\n')
            
            if line.startswith('>'):
                # Processar sequência anterior se existir
                if current_header:
                    total_sequences += 1
                    if keep_sequence:
                        fout.write(current_header + '\n')
                        fout.write('\n'.join(current_sequence) + '\n')
                        kept_sequences += 1
                
                # Nova sequência
                current_header = line
                current_sequence = []
                keep_sequence = header_contains_te(line, te_list)
            else:
                # Linha de sequência
                if line.strip():  # Ignorar linhas vazias
                    current_sequence.append(line)
        
        # Processar última sequência
        if current_header:
            total_sequences += 1
            if keep_sequence:
                fout.write(current_header + '\n')
                fout.write('\n'.join(current_sequence) + '\n')
                kept_sequences += 1
    
    return total_sequences, kept_sequences

def process_subdirectory(subdir_path, te_list):
    """
    Processa todos os arquivos FASTA em um subdiretório.
    
    Args:
        subdir_path: Path do subdiretório
        te_list: Lista de TEs para filtrar
        
    Returns:
        tuple: (total_processado, total_mantido, num_arquivos)
    """
    # Criar diretório de saída dentro do subdiretório
    output_dir = subdir_path / 'filtered'
    output_dir.mkdir(exist_ok=True)
    
    # Encontrar todos os arquivos .fasta no subdiretório
    fasta_files = list(subdir_path.glob('*.fasta'))
    
    if not fasta_files:
        return 0, 0, 0
    
    total_kept = 0
    total_processed = 0
    
    for fasta_file in fasta_files:
        # Definir arquivo de saída
        output_file = output_dir / f"{fasta_file.stem}_filter.fasta"
        
        try:
            # Filtrar arquivo
            total_seqs, kept_seqs = filter_fasta(fasta_file, output_file, te_list)
            
            percentage = (kept_seqs / total_seqs * 100) if total_seqs > 0 else 0
            
            print(f"   ✅ {fasta_file.name}")
            print(f"      Total: {total_seqs} | Mantidas: {kept_seqs} ({percentage:.1f}%)")
            
            total_kept += kept_seqs
            total_processed += total_seqs
            
        except Exception as e:
            print(f"   ❌ Erro ao processar {fasta_file.name}: {e}")
    
    return total_processed, total_kept, len(fasta_files)

def main():
    """Função principal para processar todos os subdiretórios."""
    
    print("=" * 70)
    print("🔬 FILTRO DE ELEMENTOS TRANSPONÍVEIS EM MÚLTIPLOS SUBDIRETÓRIOS")
    print("=" * 70)
    print(f"🔍 Filtrando por {len(TES)} TEs")
    print()
    
    # Estatísticas gerais
    grand_total_processed = 0
    grand_total_kept = 0
    grand_total_files = 0
    
    # Processar cada subdiretório
    for subdir_name in SUBDIRS:
        subdir_path = Path(subdir_name)
        
        print(f"📂 Processando: {subdir_name}/")
        print("-" * 70)
        
        if not subdir_path.exists():
            print(f"   ⚠️  Subdiretório não encontrado. Pulando...\n")
            continue
        
        if not subdir_path.is_dir():
            print(f"   ⚠️  Não é um diretório. Pulando...\n")
            continue
        
        total_proc, total_kept, num_files = process_subdirectory(subdir_path, TES)
        
        if num_files == 0:
            print(f"   ⚠️  Nenhum arquivo .fasta encontrado\n")
            continue
        
        percentage = (total_kept / total_proc * 100) if total_proc > 0 else 0
        
        print(f"\n   📊 Resumo {subdir_name}:")
        print(f"      Arquivos processados: {num_files}")
        print(f"      Sequências totais: {total_proc}")
        print(f"      Sequências mantidas: {total_kept} ({percentage:.1f}%)")
        print(f"      Saída: {subdir_name}/filtered/")
        print()
        
        grand_total_processed += total_proc
        grand_total_kept += total_kept
        grand_total_files += num_files
    
    # Resumo final
    print("=" * 70)
    print("📊 RESUMO GERAL DE TODOS OS SUBDIRETÓRIOS")
    print("=" * 70)
    print(f"   Total de arquivos processados: {grand_total_files}")
    print(f"   Total de sequências processadas: {grand_total_processed}")
    print(f"   Total de sequências mantidas: {grand_total_kept}")
    if grand_total_processed > 0:
        print(f"   Percentual geral mantido: {grand_total_kept/grand_total_processed*100:.1f}%")
    print()
    print("✅ Processamento completo!")

if __name__ == "__main__":
    main()
#####################################################################################################################################################

# CTRL + O = Salvar
# CTRL + X = Sair 

# Tornar o script executável 
chmod +x filter_script.py 

python3 filter_script.py
```

```python
# Separando os elementos de Transposons (Tns) e Sequências de Inserção (ISs)

## Ir para o diretório BLASTn, onde estão os diretórios Fish, Human e Bovine com os arquivos filtrados dentro da pasta filtered
cd /mnt/dados/victor_baca/outputs/size_and_density_chart/BLASTn
### Criar o diretório filtered 
mkdir filtered
cd /mnt/dados/victor_baca/outputs/size_and_density_chart/BLASTn/filtered
### Dentro do diretório filtered, criar os diretórios Fish, Human e Bovine
mkdir Fish Human Bovine

#### Agora copiamos os arquivos filtrados de cada diretório para o diretório filtered correspondente
cp /mnt/dados/victor_baca/outputs/size_and_density_chart/BLASTn/Fish/filtered/* /mnt/dados/victor_baca/outputs/size_and_density_chart/BLASTn/filtered/Fish
cp /mnt/dados/victor_baca/outputs/size_and_density_chart/BLASTn/Human/filtered/* /mnt/dados/victor_baca/outputs/size_and_density_chart/BLASTn/filtered/Human
cp /mnt/dados/victor_baca/outputs/size_and_density_chart/BLASTn/Bovine/filtered/* /mnt/dados/victor_baca/outputs/size_and_density_chart/BLASTn/filtered/Bovine

##### Agora criamos o script para separar os elementos de Transposons (Tns) e Sequências de Inserção (ISs)
nano separate_TEs.py

#####################################################################################################################################################
#!/usr/bin/env python3

import os
import re
from pathlib import Path

def create_output_directories(base_dir):
    """Cria as subpastas Tns e ISs dentro do diretório base"""
    tn_dir = os.path.join(base_dir, "Tns")
    is_dir = os.path.join(base_dir, "ISs")
    Path(tn_dir).mkdir(exist_ok=True)
    Path(is_dir).mkdir(exist_ok=True)
    return tn_dir, is_dir

def identify_element_type(header):
    """
    Identifica o tipo de elemento baseado no header.
    Retorna: 'Tn', 'IS', ou None
    """
    header = header.lstrip('>')
    
    tn_pattern = r'\b(Tn[A-Za-z0-9]+|In[_0-9])'
    is_pattern = r'\bIS[A-Za-z0-9]'
    
    if re.search(is_pattern, header, re.IGNORECASE):
        return 'IS'
    elif re.search(tn_pattern, header, re.IGNORECASE):
        return 'Tn'
    
    return None

def process_fasta_file(input_file, tn_dir, is_dir):
    """
    Processa um arquivo FASTA e separa em Tns e ISs
    """
    base_name = os.path.splitext(os.path.basename(input_file))[0]
    
    tn_output = os.path.join(tn_dir, f"{base_name}_Tns.fasta")
    is_output = os.path.join(is_dir, f"{base_name}_ISs.fasta")
    
    tn_count = 0
    is_count = 0
    other_count = 0
    other_headers = []
    
    with open(tn_output, 'w') as tn_file, \
         open(is_output, 'w') as is_file, \
         open(input_file, 'r') as infile:
        
        current_header = None
        current_sequence = []
        element_type = None
        
        for line in infile:
            line = line.rstrip('\n')
            
            if line.startswith('>'):
                if current_header and current_sequence:
                    sequence = '\n'.join(current_sequence)
                    
                    if element_type == 'Tn':
                        tn_file.write(f"{current_header}\n{sequence}\n")
                        tn_count += 1
                    elif element_type == 'IS':
                        is_file.write(f"{current_header}\n{sequence}\n")
                        is_count += 1
                    else:
                        other_count += 1
                        other_headers.append(current_header)
                
                current_header = line
                current_sequence = []
                element_type = identify_element_type(line)
                
            else:
                if line.strip():
                    current_sequence.append(line)
        
        if current_header and current_sequence:
            sequence = '\n'.join(current_sequence)
            
            if element_type == 'Tn':
                tn_file.write(f"{current_header}\n{sequence}\n")
                tn_count += 1
            elif element_type == 'IS':
                is_file.write(f"{current_header}\n{sequence}\n")
                is_count += 1
            else:
                other_count += 1
                other_headers.append(current_header)
    
    return tn_count, is_count, other_count, other_headers

def process_directory(directory):
    """
    Processa todos os arquivos FASTA em um diretório
    """
    print(f"\n{'='*60}")
    print(f"Processando diretório: {directory}")
    print(f"{'='*60}")
    
    # Cria diretórios de saída
    tn_dir, is_dir = create_output_directories(directory)
    print(f"✓ Diretórios criados: {os.path.join(directory, 'Tns/')} e {os.path.join(directory, 'ISs/')}")
    
    # Lista arquivos FASTA no diretório
    fasta_files = [f for f in os.listdir(directory) 
                   if f.endswith(('.fasta', '.fa')) and 
                   os.path.isfile(os.path.join(directory, f))]
    
    if not fasta_files:
        print(f"⚠ Nenhum arquivo FASTA encontrado em {directory}")
        return 0, 0, 0, []
    
    print(f"✓ Encontrados {len(fasta_files)} arquivo(s) FASTA\n")
    
    total_tn = 0
    total_is = 0
    total_other = 0
    all_other_headers = []
    
    for fasta_file in sorted(fasta_files):
        file_path = os.path.join(directory, fasta_file)
        print(f"  Processando: {fasta_file}")
        
        tn_count, is_count, other_count, other_headers = process_fasta_file(
            file_path, tn_dir, is_dir
        )
        
        print(f"    ├─ Transposons (Tns): {tn_count}")
        print(f"    ├─ Seq. Inserção (ISs): {is_count}")
        print(f"    └─ Outros elementos: {other_count}")
        
        if other_count > 0:
            print(f"\n    ⚠ Headers não classificados:")
            for header in other_headers[:3]:  # Mostra apenas os 3 primeiros
                print(f"       {header}")
            if len(other_headers) > 3:
                print(f"       ... e mais {len(other_headers) - 3} headers")
            all_other_headers.extend([(fasta_file, header) for header in other_headers])
        
        print()
        
        total_tn += tn_count
        total_is += is_count
        total_other += other_count
    
    print(f"{'─'*60}")
    print(f"RESUMO - {directory}")
    print(f"{'─'*60}")
    print(f"Total de Transposons (Tns): {total_tn}")
    print(f"Total de Seq. Inserção (ISs): {total_is}")
    print(f"Total de outros elementos: {total_other}")
    
    return total_tn, total_is, total_other, all_other_headers

def main():
    """Função principal"""
    print("="*60)
    print("Separador de Elementos Tns e ISs - Múltiplos Diretórios")
    print("="*60)
    
    # Diretórios a serem processados, esta lista e onde o script irá procurar os diretórios
    ## Modificar conforme necessário
    target_dirs = ['Fish', 'Human', 'Bovine']
    
    # Verifica quais diretórios existem
    existing_dirs = [d for d in target_dirs if os.path.isdir(d)]
    
    if not existing_dirs:
        print(f"\n⚠ Nenhum dos diretórios esperados foi encontrado: {', '.join(target_dirs)}")
        print("Certifique-se de executar o script no diretório 'filtered'")
        return
    
    print(f"\n✓ Diretórios encontrados: {', '.join(existing_dirs)}\n")
    
    # Totais gerais
    grand_total_tn = 0
    grand_total_is = 0
    grand_total_other = 0
    
    # Processa cada diretório
    for directory in existing_dirs:
        total_tn, total_is, total_other, other_headers = process_directory(directory)
        grand_total_tn += total_tn
        grand_total_is += total_is
        grand_total_other += total_other
    
    # Resumo final geral
    print(f"\n{'='*60}")
    print("RESUMO GERAL - TODOS OS DIRETÓRIOS")
    print(f"{'='*60}")
    print(f"Total de Transposons (Tns): {grand_total_tn}")
    print(f"Total de Seq. Inserção (ISs): {grand_total_is}")
    print(f"Total de outros elementos: {grand_total_other}")
    
    print("\n✓ Processamento concluído!")
    print(f"\nArquivos salvos em:")
    for directory in existing_dirs:
        print(f"  ├─ {directory}/Tns/")
        print(f"  └─ {directory}/ISs/")

if __name__ == "__main__":
    main()
#####################################################################################################################################################

# CTRL + O = Salvar
# CTRL + X = Sair 

chmod +x separate_TEs.py
python3 separate_TEs.py

# Por ultimo se utiliza este comando para remover arquivos vazios criados no processo
## Este comando verifica e remove todos os arquivos vazios dentro do diretório atual e seus subdiretórios
find . -type f -empty -delete
```

```python
# Rodando cd-hit-est

## Maneira manual para um arquivo
cd /mnt/dados/victor_baca/outputs/size_and_density_chart/BLASTn/filtered/Fish/ISs/
mkdir cd-hit-est
### Adicionando o CD-HIT ao PATH temporariamente 
export PATH=$PATH::/mnt/dados/victor_baca/tools/cdhit
### Rodando o cd-hit-est para um arquivo específico 
nohup cd-hit-est -i input.fasta -c 0.80 -d 0 -aS 0.80 -aL 0.80 -r 1 -G 0 -g 1 -o cd-hit-est/output.fasta &

## Maneira automatizada para todos os arquivos em múltiplos diretórios
### Ir para o diretório filtered
cd /mnt/dados/victor_baca/outputs/size_and_density_chart/BLASTn/filtered
### Criar o script run_cdhit_all.sh
nano run_cdhit_all.sh

#####################################################################################################################################################
#!/bin/bash

# Script para executar cd-hit-est em arquivos FASTA de múltiplos diretórios
# Uso: bash run_cdhit_all.sh

# Diretórios a processar
DIRS=("Fish/ISs/" "Human/ISs/" "Human/Tns/" "Bovine/ISs/" "Bovine/Tns/")

# Parâmetros do cd-hit-est
c_IDENTITY=0.80
d_Description=0
aS=0.80
aL=0.80
r=1
G=0
g=1

echo "==================================="
echo "Iniciando processamento cd-hit-est"
echo "==================================="
echo ""

# Processar cada diretório
for DIR in "${DIRS[@]}"; do
    echo "Processando diretório: $DIR"
    
    # Verificar se o diretório existe
    if [ ! -d "$DIR" ]; then
        echo "ERRO: Diretório $DIR não encontrado!"
        continue
    fi
    
    # Criar subdiretório cd-hit-est
    OUTPUT_DIR="$DIR/cd-hit-est"
    mkdir -p "$OUTPUT_DIR"
    echo "Diretório de output criado: $OUTPUT_DIR"
    
    # Contar arquivos FASTA
    FASTA_COUNT=$(ls "$DIR"/*.fasta 2>/dev/null | wc -l)
    
    if [ "$FASTA_COUNT" -eq 0 ]; then
        echo "Nenhum arquivo *.fasta encontrado em $DIR"
        echo ""
        continue
    fi
    
    echo "Encontrados $FASTA_COUNT arquivos FASTA em $DIR"
    echo ""
    
    # Processar cada arquivo FASTA
    COUNTER=0
    for INPUT_FILE in "$DIR"/*.fasta; do
        COUNTER=$((COUNTER + 1))
        
        # Extrair o nome base do arquivo
        BASENAME=$(basename "$INPUT_FILE" .fasta)
        
        # Definir arquivo de output
        OUTPUT_FILE="$OUTPUT_DIR/${BASENAME}_cdhit.fasta"
        
        echo "[$COUNTER/$FASTA_COUNT] Processando: $(basename $INPUT_FILE)"
        echo "  Output: $OUTPUT_FILE"
        
        # Executar cd-hit-est
        cd-hit-est \
            -i "$INPUT_FILE" \
            -c $c_IDENTITY \
            -d $d_Description \
            -aS $aS \
            -aL $aL \
            -r $r \
            -G $G \
            -g $g \
            -o "$OUTPUT_FILE" \
            > "$OUTPUT_FILE.log" 2>&1
        
        # Verificar se o comando foi executado com sucesso
        if [ $? -eq 0 ]; then
            echo "  ✓ Concluído com sucesso"
        else
            echo "  ✗ ERRO ao processar (verifique $OUTPUT_FILE.log)"
        fi
        echo ""
    done
    
    echo "Diretório $DIR concluído!"
    echo "-----------------------------------"
    echo ""
done

echo "==================================="
echo "Processamento finalizado!"
echo "==================================="
echo ""
echo "Resumo dos outputs:"
for DIR in "${DIRS[@]}"; do
    if [ -d "$DIR/cd-hit-est" ]; then
        COUNT=$(ls "$DIR/cd-hit-est"/*_cdhit.fasta 2>/dev/null | wc -l)
        echo "  $DIR/cd-hit-est: $COUNT arquivos processados"
    fi
done
#####################################################################################################################################################

# CTRL+O para salvar
# CTRL+X para sair

chmod +x run_cdhit_all.sh

./run_cdhit_all.sh
```

```python
cd /mnt/dados/victor_baca/outputs/size_and_density_chart/BLASTn/filtered

# Script para contar elementos de TEs em arquivos FASTA do BLASTn
nano contar_blastn.py

#####################################################################################################################################################
import re
import sys
from collections import Counter

# Verifica se o arquivo foi fornecido como argumento
if len(sys.argv) < 2:
    print("Uso: python contar_tes.py <caminho_do_arquivo.fasta>")
    print("Exemplo: python contar_tes.py /home/user/dados/Human_cdhit.fasta")
    sys.exit(1)

# Obtém o caminho do arquivo do argumento
filename = sys.argv[1]

# Dicionário para contar os TEs
te_counts = Counter()

# Verifica se o arquivo existe e processa
try:
    with open(filename, 'r') as file:
        for line in file:
            # Processa apenas as linhas de cabeçalho (começam com >)
            if line.startswith('>'):
                # Extrai o nome do TE do cabeçalho
                # Formato esperado: >blast_hit_XX ... Name=TE_NAME;...
                match = re.search(r'Name=([^;]+)', line)
                if match:
                    te_name = match.group(1)
                    te_counts[te_name] += 1

except FileNotFoundError:
    print(f"Erro: Arquivo '{filename}' não encontrado!")
    print("Verifique se o caminho está correto.")
    sys.exit(1)
except Exception as e:
    print(f"Erro ao processar o arquivo: {e}")
    sys.exit(1)

# Verifica se algum TE foi encontrado
if not te_counts:
    print("Aviso: Nenhum elemento transponível encontrado no arquivo.")
    print("Verifique se o arquivo está no formato correto.")
    sys.exit(0)

# Exibe os resultados
print("=" * 60)
print("ELEMENTOS TRANSPONÍVEIS ENCONTRADOS")
print("=" * 60)
print(f"\nTotal de TEs únicos: {len(te_counts)}")
print(f"Total de ocorrências: {sum(te_counts.values())}\n")

print("-" * 60)
print(f"{'Nome do TE':<40} {'Quantidade':>10}")
print("-" * 60)

# Ordena por quantidade (decrescente) e depois por nome
for te_name, count in sorted(te_counts.items(), key=lambda x: (-x[1], x[0])):
    print(f"{te_name:<40} {count:>10}")

print("-" * 60)

# Estatísticas adicionais
print("\n" + "=" * 60)
print("ESTATÍSTICAS")
print("=" * 60)
print(f"TE mais frequente: {te_counts.most_common(1)[0][0]} ({te_counts.most_common(1)[0][1]} ocorrências)")
print(f"Média de ocorrências por TE: {sum(te_counts.values()) / len(te_counts):.2f}")
#####################################################################################################################################################

# CTRL + O -> para salvar o arquivo
# CTRL + X -> para sair do editor

chmod +x contar_blastn.py

python contar_blastn.py Bovine/ISs/cd-hit-est/TA_B490_blastn_filter_ISs_cdhit.fasta
```
