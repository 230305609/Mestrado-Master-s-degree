### https://www.bioinformatics.org/cd-hit/

### O CD-HIT (Cluster Database at High Identity with Tolerance) é uma ferramenta bioinformática amplamente utilizada para reduzir redundância em conjuntos de dados de sequências biológicas (proteínas ou nucleotídeos). Ele agrupa sequências altamente similares em clusters, mantendo apenas uma sequência representativa de cada cluster. Isso é especialmente útil em estudos com elementos transponíveis (TEs), onde há muitas cópias repetidas ou variantes de uma mesma família.

### https://www.bioinformatics.org/cd-hit/cd-hit-user-guide

```python
# utilizar cd-hit para remover redundância dessas anotações
# cd-hit-est versão do CD-HIT para sequências nucleotídicas
cd-hit-est -i input.fasta -c 0.80 -d 0 -aS 0.80 -aL 0.80 -r 1 -G 0 -g 1 -o output.fasta

## -c 0.8  -> limiar (Threshold) de identidade de sequência (80%), Sequências com ≥80% de identidade são agrupadas no mesmo cluster

## -d 0    -> Comprimento da descrição no arquivo de saída, se não especificado como 0, a descrição pode ser truncada em 20 caracteres. Este parâmetro e opcional

## -aS 0.8 -> Cobertura mínima da sequência curta em relação à sequência longa para agrupamento (80%), ou seja, Cobertura de alinhamento da sequência MAIS CURTA, como coloquei 0.80 a sequência curta deve estar coberta em 80% pelo alinhamento

## -aL 0.8 -> Cobertura mínima da sequência longa em relação à sequência curta para agrupamento (80%), ou seja, Cobertura de alinhamento da sequência MAIS LONGA, como coloquei 0.80 a sequência longa deve estar coberta em 80% pelo alinhamento

## -r 1    -> Como coloquei 1, este parâmetro vai fazer com que seja comparada ambas as fitas (forward e reverse complement), este parâmetro se coloca, pois os TEs podem estar em orientações opostas no genoma e sem este parâmetro, você pode ter duplicatas que são reverse complement uma da outra

## -G 0    -> Este parâmetro por defaut é 1 fara alinhamento global, mas como coloquei 0, fará alinhamento local. Este parametro e colocado, pois O CD-HIT pode falhar se a cobertura da sequência mais longa (-aL) for implicitamente considerada. Este parametro e uma solução para isso.

## -g 1    -> este parâmetro com 1 ativa o modo de clustering mais preciso (mais lento, mas mais acurado). Parâmetro opcional

# Em servidores, desta forma o processo continua rodando mesmo se a conexão for perdida
nohup cd-hit-est -i input.fasta -c 0.80 -d 0 -aS 0.80 -aL 0.80 -r 1 -G 0 -g 1 -o output.fasta &

```
